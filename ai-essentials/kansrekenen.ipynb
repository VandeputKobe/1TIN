```python
from scipy.stats import norm
import math
mu = 0
s = math.sqrt(1)
```

```python
print("kans dat z < -1.23 =", norm.cdf(-1.23, mu, s))
```

```python
print("kans dat z > 2.09 =", 1 - norm.cdf(2.09, mu, s))
```

```python
print("kans dat 1.21 < z < 2.85 =", norm.cdf(2.85, mu, s) - norm.cdf(1.21, mu, s))
```

```python
print("a als p(z < a) = 0.9936", norm.ppf(0.9936, mu, s))
```

```python
print("a als p(z >= a) = 0.0281", norm.ppf(1 - 0.0281, mu, s))
```
