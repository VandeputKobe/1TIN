```python
#oef 11
df["survived"].value_counts().plot(kind="bar")
plt.show()
```

```python
#oef 12 oplossing 1
men = df[df["who"] == "man"][["who","survived"]]
woman = df[df["who"] == "woman"][["who","survived"]]

men["status"] = men["survived"].map({0: "not survived", 1: "survived"})
men["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green"])
plt.title("survived mannen")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()

woman["status"] = woman["survived"].map({0: "not survived", 1: "survived"})
woman["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green"])
plt.title("survived woman")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()
```

```python
#oef 12 oplossing 2
men = df[df["who"] == "man"][["who","survived"]]
woman = df[df["who"] == "woman"][["who","survived"]]

men["status"] = men["survived"].map({0: "not survived", 1: "survived"})
men["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="bar")
plt.title("survived mannen")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()

woman["status"] = woman["survived"].map({0: "not survived", 1: "survived"})
woman["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="bar")
plt.title("survived woman")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()
```

```python
#oef 13
df["pclass"].value_counts().reindex([1, 2, 3]).plot(kind="bar")
```

```python
#oef 14
men = df[df["who"] == "man"][["who","pclass"]]
woman = df[df["who"] == "woman"][["who","pclass"]]

men["status"] = men["pclass"].map({1: "first class", 2: "second class", 3: "third class"})
men["status"].value_counts().reindex(["first class", "second class", "third class"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green","blue"])
plt.title("pclass mannen")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()

woman["status"] = woman["pclass"].map({1: "first class", 2: "second class", 3: "third class"})
woman["status"].value_counts().reindex(["first class", "second class", "third class"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green","blue"])
plt.title("pclass woman")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()


```

```python
#oef 17
df = pd.read_csv("titanic.csv")

# boxplot van leeftijd
df["age"].plot.box(vert=False)  # horizontale boxplot

plt.title("Boxplot van leeftijd")
plt.xlabel("Leeftijd")
plt.ylabel("Waarden")
plt.show()
```

```python
#oef 18
# Boxplot van leeftijd per klasse (horizontaal)
df.boxplot(column="age", by="pclass", vert=False)

plt.title("Leeftijd per klasse")
plt.suptitle("")  # Verwijder de standaard titel die boxplot toevoegt
plt.xlabel("Leeftijd")
plt.ylabel("Klasse")
plt.show()
```
