```python
#oef 12 oplossing 1
men = df[df["who"] == "man"][["who","survived"]]
woman = df[df["who"] == "woman"][["who","survived"]]

men["status"] = men["survived"].map({0: "not survived", 1: "survived"})
men["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green"])
plt.title("survived mannen")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()

woman["status"] = woman["survived"].map({0: "not survived", 1: "survived"})
woman["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="pie",autopct="%1.1f%%",colors=["red","green"])
plt.title("survived woman")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()
```

```python
#oef 12 oplossing 2
men = df[df["who"] == "man"][["who","survived"]]
woman = df[df["who"] == "woman"][["who","survived"]]

men["status"] = men["survived"].map({0: "not survived", 1: "survived"})
men["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="bar")
plt.title("survived mannen")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()

woman["status"] = woman["survived"].map({0: "not survived", 1: "survived"})
woman["status"].value_counts().reindex(["not survived", "survived"]).plot(kind="bar")
plt.title("survived woman")
plt.xlabel("overleefd")
plt.ylabel("aantal")
plt.show()
```

```python
#oef 17
df = pd.read_csv("titanic.csv")

# boxplot van leeftijd
df["age"].plot.box(vert=False)  # horizontale boxplot

plt.title("Boxplot van leeftijd")
plt.xlabel("Leeftijd")
plt.ylabel("Waarden")
plt.show()
```

```python
#oef 18
# Boxplot van leeftijd per klasse (horizontaal)
df.boxplot(column="age", by="pclass", vert=False)

plt.title("Leeftijd per klasse")
plt.suptitle("")  # Verwijder de standaard titel die boxplot toevoegt
plt.xlabel("Leeftijd")
plt.ylabel("Klasse")
plt.show()
```

```python
#oef 24
print("gemmidelde = " + str(df["fare"].mean()))
print("mediaan = " + str(df["fare"].median()))
print("minimum = " + str(df["fare"].min()))
print("maximum = " + str(df["fare"].max()))
print("variantie = " + str(df["fare"].var()))
print("standaarddeviatie = " + str(df["fare"].std()))
print("q1 = " + str(df["fare"].quantile(0.25)))
print("q2 = " + str(df["fare"].quantile(0.50)))
print("q3 = " + str(df["fare"].quantile(0.75)))

print("modus = " + str(df["sex"].mode()))
```

```python
#oef 25
# tel aantal per klasse en geslacht
data = df.groupby(["pclass", "sex"]).size().unstack()

# lijn grafiek
plt.plot(data.index, data["male"], marker="o", label="mannen")
plt.plot(data.index, data["female"], marker="o", label="vrouwen")

plt.xlabel("Klasse (pclass)")
plt.ylabel("Aantal passagiers")
plt.xticks(data.index)  # Zorg dat x-ticks overeenkomen met klassen
plt.title("Aantal mannen en vrouwen per klasse")
plt.legend()

plt.show()
```

```python
#oef 26
# gemiddelde leeftijd per geslacht met pivot_table
gem_leeftijd = df.pivot_table(values="age", index="sex", aggfunc="mean")

print(gem_leeftijd)
```

```python
#oef 27
# Bereken gemiddelde ticketprijs per leeftijd
gem_fare_per_age = df.groupby("age")["fare"].mean()

plt.figure(figsize=(10,5))
plt.plot(gem_fare_per_age.index, gem_fare_per_age.values, marker="o", color="green")

plt.xlabel("Leeftijd (age)")
plt.ylabel("Gemiddelde ticketprijs (fare)")
plt.title("Gemiddelde ticketprijs per leeftijd")
plt.grid(True)
plt.show()
```
